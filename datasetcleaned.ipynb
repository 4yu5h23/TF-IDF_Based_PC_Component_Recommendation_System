```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('buildsfinal2.csv')
```

```python
df.head()
```

```python
cpu_counts = df['CPU'].value_counts()

# Plot the frequencies
plt.figure(figsize=(10,5))
sns.barplot(x=cpu_counts.index, y=cpu_counts.values, alpha=0.8)
plt.title('Frequency Distribution of CPUs')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('CPU', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie chart for top 5 CPUs
top_5_cpus = cpu_counts[:5]
plt.figure(figsize=(10,5))
plt.pie(top_5_cpus.values, labels=top_5_cpus.index, autopct='%1.1f%%')
plt.title('Top 5 CPUs')
plt.show()

# Scatterplot for CPU and GPU
# For this, we need to convert the CPU and GPU columns to numerical values. 
# One way to do this is by using LabelEncoder from sklearn.preprocessing.

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['CPU_encoded'] = le.fit_transform(df['CPU'])
df['GPU_encoded'] = le.fit_transform(df['GPU'])

plt.figure(figsize=(10,5))
sns.scatterplot(x='CPU_encoded', y='GPU_encoded', data=df)
plt.title('Scatterplot of CPU and GPU')
plt.xlabel('CPU')
plt.ylabel('GPU')
plt.show()
```

```python
crosstab = pd.crosstab(df['CPU'], df['GPU'])

# Plot the cross-tabulation as a heatmap
plt.figure(figsize=(10,10))
sns.heatmap(crosstab)
plt.title('Heatmap of CPU and GPU Combinations')
plt.xlabel('GPU')
plt.ylabel('CPU')
plt.show()
```

```python
# Count the frequency of each GPU
gpu_counts = df['GPU'].value_counts()

# Pie chart for top 5 GPUs
top_5_gpus = gpu_counts[:5]
plt.figure(figsize=(10,5))
plt.pie(top_5_gpus.values, labels=top_5_gpus.index, autopct='%1.1f%%')
plt.title('Top 5 GPUs')
plt.show()
```

```python
# Count the frequency of each RAM, Motherboard, and Power Supply
ram_counts = df['RAM'].value_counts()
motherboard_counts = df['Motherboard'].value_counts()
psu_counts = df['PSU'].value_counts()

# Pie chart for top 5 RAMs
top_5_rams = ram_counts[:5]
plt.figure(figsize=(10,5))
plt.pie(top_5_rams.values, labels=top_5_rams.index, autopct='%1.1f%%')
plt.title('Top 5 RAMs')
plt.show()

# Pie chart for top 5 Motherboards
top_5_motherboards = motherboard_counts[:5]
plt.figure(figsize=(10,5))
plt.pie(top_5_motherboards.values, labels=top_5_motherboards.index, autopct='%1.1f%%')
plt.title('Top 5 Motherboards')
plt.show()

# Pie chart for top 5 Power Supplies
top_5_psus = psu_counts[:5]
plt.figure(figsize=(10,5))
plt.pie(top_5_psus.values, labels=top_5_psus.index, autopct='%1.1f%%')
plt.title('Top 5 Power Supplies')
plt.show()
```

```python
# Count the frequency of each GPU
gpu_counts = df['GPU'].value_counts()

# Plot the frequencies
plt.figure(figsize=(10,5))
sns.barplot(x=gpu_counts.index, y=gpu_counts.values, alpha=0.8)
plt.title('Frequency Distribution of GPUs')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('GPU', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```
